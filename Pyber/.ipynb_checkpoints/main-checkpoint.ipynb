```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
city_data_raw=pd.read_csv("raw_data/city_data.csv")
city_data_df = pd.DataFrame({'type': city_data_raw.groupby('city').first()['type'],
                                'driver_count': city_data_raw.groupby('city').sum()['driver_count']}).reset_index()
city_data_df.head(30)
```

```python
ride_data_df=pd.read_csv("raw_data/ride_data.csv")
ride_data_df.head()
```

```python
stat_1=pd.DataFrame(ride_data_df["city"].value_counts())
stat_1.rename(columns={"city":"Total Rides"},inplace=True)
city_group=ride_data_df.groupby("city")
avg_price=city_group["fare"].mean()
stat_1["Average Fare"]=avg_price
stat_1.reset_index(inplace=True)
stat_1=stat_1.merge(city_data_df, left_on="index",right_on="city" )
stat_1.drop("index",inplace=True,axis=1)

#Setting Type as Index for easy search using LOC
stat_1.set_index("type",inplace=True)
stat_1.head()
```

```python
#print(plt.style.available)
with plt.style.context('seaborn-darkgrid'):
    plt.title("Pyber Ride Sharing Data (2016)")
    plt.xlim(0,max(stat_1["Total Rides"]+10))
    plt.xlabel("Total Numbers of Rides (Per City)")
    plt.ylabel("Average Fair ($)")
    plt.grid(color='white', linestyle='-', linewidth=0.75)
    Urban_handle = plt.scatter(stat_1.loc["Urban","Total Rides"], stat_1.loc["Urban","Average Fare"],edgecolors="black", color="orangered", label="Urban",alpha=0.5,s=stat_1.loc["Urban","driver_count"]*5)
    Sub_handle = plt.scatter(stat_1.loc["Suburban","Total Rides"], stat_1.loc["Suburban","Average Fare"],edgecolors="black", color="skyblue", label="Suburban",alpha=0.5,s=stat_1.loc["Suburban","driver_count"]*5 )
    Rural_handle = plt.scatter(stat_1.loc["Rural","Total Rides"], stat_1.loc["Rural","Average Fare"], edgecolors="black",color="yellow", label="Rural",alpha=0.5,s=stat_1.loc["Rural","driver_count"]*5)
    plt.legend(handles=[Urban_handle, Sub_handle,Rural_handle], loc="best",title="City Types")
```
